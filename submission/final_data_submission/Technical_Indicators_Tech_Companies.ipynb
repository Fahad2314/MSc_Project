```python
import pandas as pd 
import csv
import numpy as np 
import ta
```

```python
FB_df = pd.read_csv("/Users/fahad/project_repo/data/final/FB/FB.csv", index_col=0)
GOOG_df = pd.read_csv("/Users/fahad/project_repo/data/final/GOOG/GOOG.csv", index_col=0)
MAR_df = pd.read_csv("/Users/fahad/project_repo/data/final/MAR/MAR.csv", index_col=0)
PFE_df = pd.read_csv("/Users/fahad/project_repo/data/final/PFE/PFE.csv", index_col=0)
NFLX_df = pd.read_csv("/Users/fahad/project_repo/data/final/NFLX/NFLX.csv", index_col=0)
```

```python
FB_RSI=ta.momentum.rsi(FB_df['Close'],14,"False")
FB_MACD = ta.trend.macd_diff(FB_df['Close'],12,26,9,"False")
FB_ICHI_A = ta.trend.ichimoku_a(FB_df['High'],FB_df['Low'],9,26,"False","False")
FB_ICHI_B = ta.trend.ichimoku_b(FB_df['High'],FB_df['Low'],23,52,"False","False")
FB_ATR = ta.volatility.average_true_range(FB_df['High'],FB_df['Low'],FB_df['Close'],14,'False')
FB_BB_hband = ta.volatility.bollinger_hband(FB_df['Close'],20,2,"False")
FB_BB_lband = ta.volatility.bollinger_lband(FB_df['Close'],20,2,"False")
FB_BB_high_indicator = ta.volatility.bollinger_hband_indicator(FB_df['Close'],20,2,"False")
FB_BB_low_indicator = ta.volatility.bollinger_lband_indicator(FB_df['Close'],20,2,"False")
FB_df['RSI'] = FB_RSI
FB_df['MACD'] = FB_MACD
FB_df['Ichimoku_Span_A']= FB_ICHI_A
FB_df['Ichimoku_Span_B'] = FB_ICHI_B
FB_df['ATR'] = FB_ATR
FB_df['BB_High_Indicator'] = FB_BB_high_indicator
FB_df['BB_Low_Indicator'] = FB_BB_low_indicator
```
