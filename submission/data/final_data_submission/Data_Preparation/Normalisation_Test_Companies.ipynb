```python
list_1 = [FB_df,PFE_df,MAR_df,GOOG_df,NFLX_df]


list_2= ['Open','High','Low','Close','Volume','RSI','MACD','Ichimoku_Span_A','Ichimoku_Span_B','ATR']

def normalize_col(df_col):
    dataNorm=((df_col-df_col.min())/(df_col.max()-df_col.min()))
    return dataNorm


for dataframe in list_1:
    for col in list_2:
        dataframe[col] = normalize_col(dataframe[col])
```
