```python
#function takes a dataframe with OHLCV and DATETIME object as the INDEX in a %Y-%M-%D format
def window_split_data(dataframe_to_split, n_year_in_window,stock_str):
    """the function splits a monthly dataframe into 5 year windows that are non-overlapping"""
    
    #returns dataframe object with years present in the original dataframe as the index
    set_of_years = set()
    
    #print (dataframe_to_split.head(1))
    
    #print (type(dataframe_to_split.index))
    
    dates = dataframe_to_split.index
    
    dataframe_to_split['Date'] = dates
    
    dataframe_to_split['Date'] = pd.to_datetime(dataframe_to_split['Date'], format="%Y-%m-%d")
    
#     print (dataframe_to_split.head(1))
#     print (type(dataframe_to_split['Date']))
#     print (dataframe_to_split.dtypes)
    
    for timestamp in list(dataframe_to_split.Date):
        #print (timestamp.year)
        set_of_years.add(timestamp.year)
        
    #print (set_of_years)
        
    

    list_of_years_in_df = list(set_of_years)

# #   #number of training test sets we will have
    len_of_list = len(list_of_years_in_df)  
    
    x = np.ceil(len_of_list/n_year_in_window)
    
    final_year_list = []
    
    for i in range(len(list_of_years_in_df)):
        if (i%x == 0.0):
            final_year_list.append(list_of_years_in_df[i])

        else:
            pass
    
#     print ("\n")

    print (final_year_list)
    
#     print ("\n")
    
    for year in final_year_list:
        if year <=2013:
            stop = year + n_year_in_window
            start = year
            bottom = str(start)
            
            stop = str(stop)
            start = str(start)
            
            start_date = start + "-01-01"
            stop_date = stop + "-01-01"
            
            
            temp_df = dataframe_to_split.loc[(dataframe_to_split['Date'] > pd.to_datetime(start_date, format="%Y-%m-%d")) &
                                             (dataframe_to_split['Date'] < pd.to_datetime(stop_date, format="%Y-%m-%d"))]
        
            
            temp_df = temp_df.iloc[::-1]


            
            name = "{}_{}".format(bottom,stock_str)
            temp_df.to_csv("/Users/fahad/project_repo/data/final/{}/{}.csv".format(stock_str, name))
            
        elif (year==2014) or (year ==2015):
            
            stop = 2019
            
            start = year
            bottom = str(start)
            
            start_date = str(start) + "-01-01"
            stop_date = str(stop) + "-07-01"
            
            temp_df = dataframe_to_split.loc[(dataframe_to_split['Date'] > pd.to_datetime(start_date, format="%Y-%m-%d")) &
                                             (dataframe_to_split['Date'] < pd.to_datetime(stop_date, format="%Y-%m-%d"))]
                        
            
            temp_df = temp_df.iloc[::-1]
            


            name = "{}_{}".format(bottom,stock_str)
            temp_df.to_csv("/Users/fahad/project_repo/data/final/{}/{}.csv".format(stock_str, name))


        else:
            pass


```
