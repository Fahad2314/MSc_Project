```python
class Buy_and_Hold(Strategy):
    
    # Define the two MA lags as *class variables*
    # for later optimization
    n1 = 10
    n2 = 20
    
    
    def init(self):
        # Precompute two moving averages
        self.sma1 = self.I(SMA, self.data.Close, self.n1)
        self.sma2 = self.I(SMA, self.data.Close, self.n2)
        
        self.first_index = pd.to_datetime("2004-09-17", format="%Y-%m-%d")
        
    def next(self):
        
        if self.first_index == self.data.index[-1]:
            self.buy()
            
        else:
            pass
        

        
```

```python
bt = Backtest(GOOG, Buy_and_Hold, cash=10000, commission=.01)
bt.run()
```

```python
bt.plot(superimpose=False, relative_equity=True, plot_volume=False, plot_pl=False, plot_drawdown=False)
```
