```python
class Binary_Classification_SVM(Strategy):
    
    # Define the two MA lags as *class variables*
    # for later optimization
    n1= 10
    n2 = 20
    
    def init(self):
        
        self.data.Prediction
#         # Precompute two moving averages
#         self.sma1 = self.I(SMA, self.data.Close, self.n1)
#         self.sma2 = self.I(SMA, self.data.Close, self.n2)
    
        
        
#         print (self.sma1)
#         print (self.sma2)
    
    def next(self):
        # If sma1 crosses above sma2, buy the asset
        if self.data.Prediction[-1] == 1.0:
#             print("\n")
#             print (self.data.Prediction[-1])
            
            self.buy()

        # Else, if sma1 crosses below sma2, sell it
        elif self.data.Prediction[-1] == 0.0:
#             print("\n")
#             print (self.data.Prediction[-1])
            
            self.sell()
```

```python
bt = Backtest(MSFT_final_df, Binary_Classification_SVM, cash=10000, commission=.002)
bt.run()

```
